```python
#to add a new row to existing array r_() is used

i=np.array([[1, 2, 3], [45, 4, 7], [9, 6, 10]])
na=np.array([12,23,45])
new=np.r_[i,[na]]
```

```python
#To handle missing data in data frame by using "np.nan".

d={"A":[1,2,np.nan],"B":[5,np.nan,np.nan],"C":[1,2,3],"D":[np.nan,np.nan,np.nan]}

df1=pd.DataFrame(d)

df1
```

```python
#To drop the values with Nan
#dropna() is used

df1.dropna()
```

```python
#dropna(how,thresh,axis,inplace)      syntax -->dropna(parameters)

#how='all' or 'any'

print(df1.dropna(how='all'),"\n\n")   #It will be checking the rows with all Nan values to delete that row

print(df1.dropna(how='any'))  #It will delete the rows which have even single Nan value
```

```python
df1.dropna(how='all',axis=1)
```

```python
df1.isna()  #it returns true if it has Nan
```

```python
#To check the sum of Nan values in each column

df1.isna().sum()
```

```python
#thresh 

#it should have non-Nan value more than or equal to thresh value

df1.dropna(thresh=2)
```

```python
df1.dropna(thresh=2,axis=1)
```

```python
#fillna()

df1.fillna(value=0.0)   #or df1.fillna(0)
```

```python
#printing of average of values in a column and storing it in that Nan values of same column

df1['A'].fillna(value=df1['A'].mean())
```

```python
import numpy as np
x_years=np.arange(2000,2005,dtype='int')
y_innova=np.array([20,40,10,14,23])
y_fortuner=np.array([10,20,40,35,39])
plt.plot(x_years,y_innova,'-*b')
plt.plot(x_years,y_fortuner,'--+g')
plt.xlabel("years")
plt.ylabel("number of car")
plt.title('Car Statistics')
plt.legend(['Innova','Fortuner'])
plt.figure(figsize=[5,5])
```

```python
#Barplot 
products=['TV','AC','Mobile','Laptop','SmartWatch']
sales=np.random.randint(20,100,5)
plt.xlabel('Products')
plt.ylabel('No. of products sold')
plt.bar(products,sales,color=["blue","orange","red","skyblue","green"],width=0.4,edgecolor='black')
plt.title('Electronic Gadgets Statistics')
plt.figure(figsize=[6,6])
plt.xlabel('Products')
plt.ylabel('No. of products sold')
plt.barh(products,sales,color=["blue","orange","red","skyblue","green"],height=0.4,edgecolor='black')
plt.title('Electronic Gadgets Statistics')
```
